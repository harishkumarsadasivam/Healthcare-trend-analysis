```python
df = df.sort_values("date").set_index("date")
```

```python
daily_trends = df.resample("D").agg({"Glucose": "mean", "BMI": "mean", "risk_score": "mean", "Outcome": "mean", "high_glucose_flag": "mean"})
daily_trends.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
daily_trends["Glucose"].plot()
plt.title("Daily Average Glucose Trend")
plt.ylabel("Glucose")
plt.xlabel("Date")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
daily_trends["risk_score"].plot()
plt.title("Daily Risk Score Trend")
plt.ylabel("Risk Score")
plt.xlabel("Date")
plt.show()
```

```python
monthly_trends = df.resample("M").agg({"Glucose": "mean", "BMI": "mean", "risk_score": "mean", "Outcome": "mean", "high_glucose_flag": "mean"})
monthly_trends.head()
```

```python
plt.figure(figsize=(12, 5))
monthly_trends["Outcome"].plot(marker="o")
plt.title("Monthly Diabetes Rate (Outcome = 1)")
plt.ylabel("Rate")
plt.xlabel("Month")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
monthly_trends["high_glucose_flag"].plot(marker="o", color="red")
plt.title("Monthly High Glucose Alert Rate")
plt.ylabel("Alert Rate")
plt.xlabel("Month")
plt.show()
```

```python
df["risk_bucket"] = pd.qcut(df["risk_score"], q=4, labels=["Low", "Medium", "High", "Very High"])
df["risk_bucket"].value_counts()
```

```python
risk_outcome = df.groupby("risk_bucket")["Outcome"].mean().sort_index()
risk_outcome
```

```python
risk_outcome.plot(kind="bar", figsize=(8, 5))
plt.title("Diabetes Rate by Risk Bucket")
plt.ylabel("Diabetes Rate")
plt.xlabel("Risk Category")
plt.show()
```

```python
daily_trends.to_csv("../../data/processed/daily_trends.csv")
monthly_trends.to_csv("../../data/processed/monthly_trends.csv")

print("Trend tables exported successfully!")
```
